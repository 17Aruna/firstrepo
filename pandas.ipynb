```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aruna', 'Lavanya', 'ram', 'shyam']
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [23, 34, 25, 40]
})
```

```python
inner_join = pd.merge(df1, df2, on='ID', how='inner')
print("inner_join:",inner_join)
```

```python
left_join = pd.merge(df1, df2, on='ID', how='left')
print('left_join:',left_join)
```

```python
right_join = pd.merge(df1, df2, on='ID', how='right')
print('right_join:',right_join)
```

```python
outer_join = pd.merge(df1, df2, on='ID', how='outer')
print('outer_join:',outer_join)
```

```python
row_concat = pd.concat([df1, df2], axis=0, ignore_index=True)
print("row_concat:",row_concat)
```

```python
col_concat = pd.concat([df1, df2], axis=1)
print("col_concat:",col_concat)
```

```python
#loc is label_based and includes the stop in slicing
row_loc = df1.loc[1]  
name_loc = df1.loc[1, 'Name'] 
print("row_loc:",row_loc)
print("name_loc:",name_loc)
```

```python
#iloc is index -basedand excludes the stop in slicing
row_iloc = df1.iloc[1]
name_iloc = df1.iloc[1, 1] 
print("row_loc:",row_loc)
print("name_loc:",name_loc)

```
